```python
#Import libraries
from pyspark.sql import SparkSession
import great_expectations as ge
from great_expectations.dataset import PandasDataset
from datetime import datetime
from great_expectations.data_context import DataContext
```

```python
#Initialize Spark Session
spark = SparkSession.builder \
    .appName("Data Quality Pipeline") \
    .getOrCreate()
```

```python
# Read file CSV
df = spark.read.csv("test.csv", header=True, inferSchema=True)
```

```python
df.head()
```

```python
pandas_df = df.toPandas()
```

```python
dataset = PandasDataset(pandas_df)
```

```python
dataset.expect_table_row_count_to_equal(1000)
```

```python
df.printSchema()
```

```python
dataset.expect_column_to_exist('ID')
```

```python
dataset.expect_column_values_to_be_unique('ID')
```

```python
dataset.expect_column_values_to_not_be_null('ID')
```

```python
dataset.expect_column_values_to_be_in_set("Type_of_order", [ 'Buffet ', 'Meal ', 'Drinks ', 'Snack '])
```

```python
# "Define expectations for the column 'Type_of_order'
expectation_suite = dataset.expect_column_values_to_be_in_set(column="Type_of_order", value_set=['Buffet', 'Meal', 'Drinks', 'Snack'])
```

```python

results = dataset.validate(expectation_suite)
```
